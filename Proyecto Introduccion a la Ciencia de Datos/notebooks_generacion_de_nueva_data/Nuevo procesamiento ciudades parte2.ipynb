```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
path_DV2 = '/content/drive/MyDrive/Proyecto Introduccion a la Ciencia de Datos/data/map.geojson'
posicion_sect = gpd.read_file(path_DV2)
```

```python
def find_sector(x):
  if(x['latitud']!='Nada'):
    pos_frame = posicion_sect.loc[posicion_sect['geometry'].contains(Point(float(x['longitud']),float(x['latitud'])))]
    if pos_frame.shape[0] == 1:
      return pos_frame.reset_index()['Name'][0]
    else:
      return 'Desconocido'
  else:
    return 'Desconocido'
```

```python
dfGYE['Nombre_sector'] = dfGYE.apply(find_sector,axis=1)
```

```python
dfQUI['Nombre_sector'] = dfQUI.apply(find_sector,axis=1)
```

```python
for k in Lista_sectores:
  numS = datos_nulosG['Descripcion_1'].str.contains(k,case = False).sum()
  if(numS > 0):
    aux1 = datos_nulosG.loc[datos_nulosG['title'].str.contains(k,case = False)]
    aux1['Nombre_sector'] = k
    if(df_nulosG.shape[0]==0):     
      df_nulosG = aux1
    else:
      df_nulosG = pd.concat([df_nulosG,aux1])
  else:
    numS2 = datos_nulosG['Descripcion_2'].str.contains(k,case = False).sum()
    if(numS2 > 0):
      aux2 = datos_nulosG.loc[datos_nulosG['Descripcion_2'].str.contains(k,case = False)]
      aux2['Nombre_sector'] = k
      if(df_nulosG.shape[0]==0):     
        df_nulosG = aux2
      else:
        df_nulosG = pd.concat([df_nulosG,aux2])
```
