### m√©tricas

```python
df = ventas.groupby(['l3', 'Nombre_sector'], as_index=False)['precioxmt2'].mean()
```

```python
df
```

```python
data_sectores['CCMC'] = data_sectores['N centros Comerciales']/data_sectores['AreaMT2']
```

```python
df = df.merge(data_sectores[['Nombre_sector', 'CCMC']])
```

```python
df
```

```python
alq_prom = alquiler.groupby('Nombre_sector', as_index=False)['price'].mean()
alq_prom.columns = ['Nombre_sector', 'alq_prom']
```

```python
df = df.merge(alq_prom, how='left')
```

```python
df
```

```python
geo_sect['centro_sector'] = geo_sect['geometry'].apply(lambda x: x.centroid)
```

```python
geo_sect = geo_sect.merge(df[['l3', 'Nombre_sector']], left_on='Name', right_on='Nombre_sector')
```

```python
municipio_gye = Point(-79.880824863767, -2.1952322152143777)
plaza_grande_uio = Point(-78.51207140183388, -0.2200787024529181)
```

```python
def centro_ciudad(x):
  if x == 'Quito':
    return plaza_grande_uio
  elif x == 'Guayaquil':
    return municipio_gye
```

```python
geo_sect['centro_ciudad'] = geo_sect['l3'].apply(centro_ciudad)
```

```python
geo_sect['DSC'] = geo_sect.apply(lambda x: x['centro_sector'].distance(x['centro_ciudad']), axis=1)
```

```python
geo_sect
```

```python
df = df.merge(geo_sect[['Nombre_sector', 'DSC']])
```

```python
df
```

```python
df.drop_duplicates(subset = 'Nombre_sector', inplace=True)
```

```python
df['alq_prom'].fillna(0, inplace=True)
```

```python
df
```

```python
vent = ventas.drop(columns=['precioxmt2'])
```

```python
properties = pd.concat([vent, alquiler])
```

```python
results = properties.groupby('Nombre_sector')['id'].count().sort_values(ascending=False).reset_index()
```

```python
results
```

```python
idx = results['id'] > 500
results[idx]
```

```python
bath_mean = properties.groupby('Nombre_sector')['bathrooms'].mean().reset_index()
```

```python
df = df.merge(bath_mean)
```

```python
idx = properties['property_type'].isin(['Oficina', 'Local comercial'])
actividad_comercial = properties[idx].groupby('Nombre_sector')['id'].count().reset_index()
```

```python
df = df.merge(actividad_comercial, how='left')
```

```python
df['id'].fillna(0, inplace=True)
```
