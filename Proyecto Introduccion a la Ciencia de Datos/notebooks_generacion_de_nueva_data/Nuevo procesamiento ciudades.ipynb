```python
f = open('/content/drive/MyDrive/Proyecto Introduccion a la Ciencia de Datos/data/archivoGuayaquil.txt', "r",encoding="Windows-1252")
esp_li = []
dicc_GYE = {}
for linea in f:
    b = linea.rstrip().split(';')
    #esp_li.append(b)
    sector = []
    precio = []
    metroscua = []
    otros = []
    ubicacion = []
    
    sector.append(b[1])
    sector.append(b[2])

    
    for i in range(3,10):
       if(len(b)>=i+1):
        if('$' in b[i]):
          precio.append(b[i])
        elif('m²' in b[i]):
          metroscua.append(b[i])
        elif('https' in b[i]):
          ubicacion.append(b[i])
        else:
          otros.append(b[i])


    
    
    dicc_GYE[b[0]] = { 'sector' : sector, 'precio' : precio, 'metroscua' : metroscua, 'otros': otros , 'ubicacion' : ubicacion}
    

    
    esp_li.append(b[2])
f.close()
print(dicc_GYE)
#print(set(esp_li))
#df_Guayaquil = pd.read_csv(f'{path}/archivoGuayaquil.txt',encoding = "Windows-1252")
```

```python
Descripcion1G = []
Descripcion2G = []
precioG = []
mt2G = []
ubicacionG = []
for gh in dicc_GYE:
  Descripcion1G.append(dicc_GYE[gh]['sector'][0]) 
  Descripcion2G.append(dicc_GYE[gh]['sector'][1])
  
  if(len(dicc_GYE[gh]['precio'])!=0):
    precioG.append(dicc_GYE[gh]['precio'][0])
  else:
    precioG.append('Nada')
  if(len(dicc_GYE[gh]['ubicacion'])!=0):
    ubicacionG.append(dicc_GYE[gh]['ubicacion'][0])
  else:
    ubicacionG.append('Nada')
  mt2G.append(dicc_GYE[gh]['metroscua'])
```

```python
Descripcion1Q = []
Descripcion2Q = []
precioQ = []
mt2Q = []
ubicacionQ = []
for gh2 in dicc_QUI:
  Descripcion1Q.append(dicc_QUI[gh2]['sector'][0]) 
  Descripcion2Q.append(dicc_QUI[gh2]['sector'][1])
  if(len(dicc_QUI[gh2]['precio'])!=0):
    precioQ.append(dicc_QUI[gh2]['precio'][0])
  else:
    precioQ.append('Nada')
  if(len(dicc_QUI[gh2]['ubicacion'])!=0):
    ubicacionQ.append(dicc_QUI[gh2]['ubicacion'][0])
  else:
    ubicacionQ.append('Nada')
  mt2Q.append(dicc_QUI[gh2]['metroscua'])
```

```python
dfGYE = pd.DataFrame()
dfQUI = pd.DataFrame()

dfGYE['Descripcion_1'] = Descripcion1G
dfGYE['Descripcion_2'] = Descripcion2G
dfGYE['precio'] = precioG
dfGYE['mt2'] = mt2G
dfGYE['ubicacion'] = ubicacionG


dfQUI['Descripcion_1'] = Descripcion1Q
dfQUI['Descripcion_2'] = Descripcion1Q
dfQUI['precio'] = precioQ
dfQUI['mt2'] = mt2Q
dfQUI['ubicacion'] = ubicacionQ
```
