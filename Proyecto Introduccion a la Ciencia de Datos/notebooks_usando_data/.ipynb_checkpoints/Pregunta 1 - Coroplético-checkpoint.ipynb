```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Polygon
from geopandas import GeoDataFrame
```

## carga del dataset

```python
df = pd.read_csv(path)
df.drop(columns=['Unnamed: 0'], inplace =True)
df.rename(columns={'Sector':'Zona'},inplace=True)
```

```python
df3 = pd.read_csv(path3)
```

```python
posicion_sect = gpd.read_file(path_DV2)
posicion_sect['Nombre_sector'] = posicion_sect['Name']
posicion_sect.drop(columns=['Name'], inplace =True)
posicion_sect.drop_duplicates(subset=['Nombre_sector'],inplace=True)
```


### determinando a qué sector pertenece cada propiedad

```python
df = df.merge(posicion_sect,on='Nombre_sector',how='left')
```

```python
df3.rename(columns={'sector':'Nombre_sector'},inplace=True)
```

```python
nuevo_df= df3.merge(df,on='Nombre_sector',how='left')
```

###*¿Hay una diferencia significativa en los precios por metros cuadrados de las viviendas de Guayaquil y Quito, según la orientación cardinal en la que se encuentren?

```python
gye_sect=nuevo_df.loc[nuevo_df['Ciudad']=='Guayaquil']
df_gye = gye_sect.groupby('Zona')['pmc'].mean().reset_index()
df_gye.rename(columns={'pmc':'pmc_zonal'},inplace=True)
gye_sect2 = gye_sect.merge(df_gye,on='Zona',how='left')
df_gye2 = GeoDataFrame(gye_sect2)
df_gye.sort_values(by='pmc_zonal',ascending=False,inplace=True)
```

```python
# Control del tamaño de la figura del mapa
fig, ax = plt.subplots(1,2,figsize=(20, 10))
_ = fig.suptitle('Precio por metro cuadrado promedio de las viviendas en Guayaquil según la orientación cardinal donde se encuentran', fontdict={'fontsize':20, 'color': '#4873ab'})
#Barras
sns.set_theme(style="darkgrid")
_ = sns.barplot(data=df_gye, x="Zona", y="pmc_zonal", ax = ax[0])
ax[0].set_xlabel('Orientacion cardinal')
_ = ax[0].set_ylabel('Precio por metro cuadrado')

# Control del título y los ejes
ax[1].set_title('Orientacion cardinal', 
             pad = 20, 
             fontdict={'fontsize':15})
ax[1].set_xlabel('Longitud')
_ = ax[1].set_ylabel('Latitud')
 
# Mostrar el mapa finalizado
_ = df_gye2.plot(column='pmc_zonal', cmap='Paired', ax=ax[1], zorder=5, legend=True)
```

```python
uio_sect=nuevo_df.loc[nuevo_df['Ciudad']=='Quito']
df_uio = uio_sect.groupby('Zona')['pmc'].mean().reset_index()
df_uio.rename(columns={'pmc':'pmc_zonal'},inplace=True)
uio_sect2 = uio_sect.merge(df_uio,on='Zona',how='left')
df_uio2 = GeoDataFrame(uio_sect2)
df_uio.sort_values(by='pmc_zonal',ascending=False,inplace=True)
```

```python
# Control del tamaño de la figura del mapa
fig, ax = plt.subplots(1,2,figsize=(20, 10))
_ = fig.suptitle('Precio por metro cuadrado promedio de las viviendas en Quito según la orientación cardinal donde se encuentran', fontdict={'fontsize':20, 'color': '#4873ab'})
#Barras
sns.set_theme(style="darkgrid")
_ = sns.barplot(data=df_uio, x="Zona", y="pmc_zonal", ax = ax[0])
ax[0].set_xlabel('Orientacion cardinal')
_ = ax[0].set_ylabel('Precio por metro cuadrado')

# Control del título y los ejes
ax[1].set_title('Orientacion cardinal', 
             pad = 20, 
             fontdict={'fontsize':15})
ax[1].set_xlabel('Longitud')
_ = ax[1].set_ylabel('Latitud')
 
# Mostrar el mapa finalizado
_ = df_uio2.plot(column='pmc_zonal', cmap='Paired', ax=ax[1], zorder=5, legend=True)
```
